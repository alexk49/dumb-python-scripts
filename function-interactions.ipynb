```python
def player_guess():
    guess=''
    while guess not in ['0','1','2']:    
        guess = input("Pick a number: 0,1, or 2")
    return int(guess)
```

```python
def check_guess(mylist,guess):
    if mylist[guess] == 'O':
        print("Correct!")
    else:
        print("Wrong guess!")
        print(mylist)
```

```python
# initial list
mylist = [' ','O',' ']
# shuffle list 
mixedup_list = shuffle_list(mylist)
# user guess 
guess = player_guess()
# check guess 
check_guess(mixedup_list,guess)
```
