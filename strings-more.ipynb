All to lower:

```python
s.lower()
```

```python
s.count('o')
```

Strings have some methods that are like regular expressions.

s.split returns as list:

```python
s.split('e')
```

This splits everything before that e and then everything after that e into seperate lists, and will do this after every occurence

```python
s = 'hihihihihihihihihihih'
```

```python
s.split('i')
```

Partition will search for seperator and return the part before the seperator and then the part after it:

```python
s.partition('i')
```

^ this just uses the first instance, and keeps the seperator character
