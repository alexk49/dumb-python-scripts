To view received emails with python we can use the built in imaplib and email libraries in python

the imaplib library has a special syntax for searching your inbox

See lecture notes for table of keywords

```python
import imaplib
```

```python
M = imaplib.IMAP4_SSL('imap.gmail.com')
```

```python
import getpass
```

```python
email = getpass.getpass("Email: ")
password = getpass.getpass("Password: ")
```

```python
M.login(email,password)
```

```python
M.list()
```

```python
M.select('inbox')
```

```python
typ, data = M.search(None,'SUBJECT "TEST TEST TEST CAT"')
```

```python
typ
```

```python
data
```

```python
data[0]
```

```python
email_id = data[0]
```

```python
result , email_data = M.fetch(email_id,'(RFC822)')
```

```python
email_data
```

```python
raw_email = email_data[0][1]
```

```python
raw_email
```

```python
raw_email_string = raw_email.decode('utf-8')
```

```python
import email
```

```python
email_message = email.message_from_string(raw_email_string)
```

```python
for part in email_message.walk():
    
    if part.get_content_type() == 'text/plain':
        body = part.get_payload(decode=True)
        print(body)
```
