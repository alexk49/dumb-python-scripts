Style vs layout - don't forget to look at full notebook for more detail

Lots of the terms used in this are used in html / css

The styling is usually based on boostrap

```python
import ipywidgets as widgets
```

```python
from IPython.display import display
```

```python
w = widgets.IntSlider()
display(w)
```

```python
w.layout.margin = 'auto'
w.layout.height = '85px'
```

```python
x = widgets.IntSlider(value=15,description='New Slider')
display(x)
```

```python
x.layout = w.layout
```

```python
widgets.Button(description='Ordinary Button',button_style='danger')
```

```python
widgets.Button(description='Ordinary Button',button_style='info')
```

```python
b1 = widgets.Button(description='Custom Color')
b1.style.button_color = 'lightgreen'
b1
```

```python
b1.style.keys
```

```python
b2 = widgets.Button(description='NEW')
b2.style = b1.style
b2
```

```python
s1 = widgets.IntSlider(description='My handle')
s1.style.handle_color = 'blue'
s1
```

```python
s1.style.keys
```
