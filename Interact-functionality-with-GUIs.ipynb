There are lots of libraries for GUIs but this is just for jupyter

```python
from ipywidgets import interact,interactive,fixed
import ipywidgets as widgets
```

```python
def func(x):
    return x
```

```python
interact(func,x=10)
```

```python
def func1(x):
    return x
```

```python
interact(func1,x=True)
```

```python
def func2(x):
    return x
```

```python
interact(func2,x='hello')
```

Using decorator:

```python
@interact(x=True,y=1.0)
def g(x,y):
    return (x,y)
```

```python
@interact(x=True,y=fixed(1.0))
def g2(x,y):
    return (x,y)
```

```python
@interact(x=True,y=fixed('hello'))
def g3(x,y):
    return (x,y)
```

```python
interact(func,x=1)
```

```python
interact(func,x=10)
```

Min is defined by -x and max is defined by x times 3

This can be changed by defining min/maxs:

```python
interact(func,x=widgets.IntSlider(min-100,max=100,step=1,value=0))
```

```python
def f(x):
    return x
```

```python
interact(f, x=widgets.IntSlider(min=-10,max=30,step=1,value=10))
```

```python
interact(func,x=(-100,100,1))
```

```python
@interact(x=(0.0,20.0,0.5))
def h(x=5.0):
    return x
```

```python
interact(func,x=['hello','option 2','option 3'])
```

```python
interact(func,x={'one':10,'two':20})
```
