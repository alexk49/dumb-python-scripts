Intersection method, will return the intersection as a new set - which is the elements that are common to all of the sets

```python
s1 = {1,2,3}
```

```python
s2 = {1,2,4}
```

```python
s1.intersection(s2)
```

Intersection update will update set with the intersection of itself and another:

```python
s1.intersection_update(s2)
```

```python
s1
```

s1 is now equal to the intersection values of the two sets
