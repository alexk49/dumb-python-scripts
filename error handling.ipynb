```python
try:
    f = open('testfile','r')
    f.write("Write a test line")
except TypeError:
    print("There was a type error!")
except OSError:
    print("Hey you have an OS Error")
except:
    print("All other exceptions!")
finally:
    # will always execute no matter what
    print("I always run")
```

```python
def ask_for_int():
    try:
        result = int(input("Please provide a number: "))
    except:
        print("Whoops! That is not a number")
    finally:
        print("End of try/except/finally")
```

```python
ask_for_int()
```

```python
def ask_for_int2():
    while True:
        
        try:
            result = int(input("Please provide a number: "))
        except:
            print("Whoops! That is not a number")
            continue
        else:
            print("Yes, thank you")
            break
        finally:
            print("End of try/except/finally")
            print("I will always run at the end")
```

```python
ask_for_int2()
```

```python
def ask():
    while True:
        try:
            num = int(input("Please enter an integer: "))
        except:
            print("That's not a number")
            continue
        else:
            answer = (num**2)
            print("Thank you, your number squared is: %d" % (answer))
            break
```

```python
ask()
```
