built in smtplib library should do a lot of the steps of connecting to an email server, confirming connection, setting a protocol, loggin on , and sending the message. 

Each major email provider has their own SMTP - simplpe mail transfer protocol server

eg. gmail = smtp.gmail.com

you have to generate an app password for gmail that basically tells them that you are going to connect to gmail via a different application

## Sending emails

```python
import smtplib
```

```python
smtp_object = smtplib.SMTP('smtp.gmail.com',587)
```

```python
smtp_object.ehlo()
```

```python
smtp_object.starttls()
```

```python
import getpass

# note that for gmail you have to generate an app password - https://myaccount.google.com/apppassword and this is what you enter in password 
```

```python
password = getpass.getpass('Password please: ')
```

```python
email = getpass.getpass("Email: ")
password = getpass.getpass("Password: ")
smtp_object.login(email,password)
```

```python
from_address = email
to_address = email
subject = input("Enter the subject line: ")
message = input("Enter the body message: ")
msg = "Subject: "+subject+'\n'+message

smtp_object.sendmail(from_address,to_address,msg)
```

```python
smtp_object.quit()
```
