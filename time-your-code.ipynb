multiple solutions, so good way to check what is the most efficient

either:

- track time elapsed

- timeit module (built in to python, is the most common) 

- %%timeit for jupyter 

```python
def func_one(n):
    return [str(num) for num in range(n)]
```

```python
func_one(10)
```

```python
def func_two(n):
    return list(map(str,range(n)))
```

```python
func_two(10)
```

```python
import time
```

```python
# current time before
start_time = time.time()
# run code
result = func_one(1000000)
# current time after running code
end_time = time.time()
# take difference for elapsed time
elapsed_time = end_time - start_time

print(elapsed_time)
```

```python
# current time before
start_time = time.time()
# run code
result = func_two(1000000)
# current time after running code
end_time = time.time()
# take difference for elapsed time
elapsed_time = end_time - start_time

print(elapsed_time)
```

```python
import timeit
```

```python
stmt = '''
func_one(100)
'''
```

```python
setup = '''
def func_one(n):
    return [str(num) for num in range(n)]
'''
```

```python
timeit.timeit(stmt,setup,number=10000000)
```

```python
stmt2 = '''
func_two(100)
'''
```

```python
setup2 = '''
def func_two(n):
    return list(map(str,range(n)))
'''
```

```python
timeit.timeit(stmt2,setup2,number=10000000)
```

```python
%%timeit
func_one(100)
```

```python
%%timeit
func_two(200)
```
