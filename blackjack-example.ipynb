```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

playing = True
```

```python
class Card():
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
class Deck():
# deck left as self, as it is a standardised thing
#should be the same every time 
    def __init__(self):
        self.deck = [] 
        for suit in suits:
            for rank in ranks:
    #goes thru for loop appending list, with card for each suit and rank
                self.deck.append(Card(suit,rank))
    
    def __str__(self):
        #empty str
        deck_comp = ''
        for card in self.deck:
            #loop through cards in deck, call string method so they print readably
            deck_comp += '\n'+ card.__str__()
        return "The deck has: " + deck_comp
    
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        single_card = self.deck.pop()
        return single_card
    
                
```

```python
test_deck = Deck()
test_deck.shuffle()
print(test_deck)
```

```python
class Hand():
    
    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0
        
    def add_card(self,card):
    # card passed in from Deck.deal()
    # single Card(suit,rank)
        self.cards.append(card)
        self.value += values[card.rank]
        
        # count aces with each new card
        if card.rank == 'Ace':
            self.aces += 1
        
    def adjust_for_ace(self):
        # ace is first considered to be 11
        # if total value > 21 and still have an ace
        # than change my ace to be a 1 instead of an 11
        # self.aces returns false unless, it has a value
        while self.value > 21 and self.aces:
        
            self.value -= 10
            self.aces -= 1
                
```

```python
class Chips:
    
    def __init__(self):
        self.total = 100
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    #take in chips
    
    while True:
        
        try:
            chips.bet = int(input("How many chips would you like to bet? "))
            
        except ValueError:
            print("Sorry, please provide an integer")
            
        else:
            if chips.bet > chips.total:
                print("Sorry you do not have enough chips. You have {} left".format(chips.total))
                continue
            else:
                break
```

```python
def hit(deck,hand):
    single_card = deck.deal()
    hand.add_card(single_card)
    hand.adjust_for_ace()
```

```python
def hit_or_stand(deck,hand):
    global playing

    x = input("Hit or Stand? Enter h or so ")
        
    if x[0].lower() == 'h':
        hit(deck,hand)
    elif x[0].lower() == 's':
        print("Player stands. Dealer's turn")
        playing = False
            
    else:
        print("Sorry, I did not understand. Please enter h or so only.")
```

```python
def show_some(player,dealer):
    # show only one of the dealer's cards
    print("\n Dealer's hand: ")
    print("First card hidden")
    print(dealer.cards[1])
    # show all (2 cards) of the player's hand/cards
    
    print("\n Player's hand:")
    for card in player.cards:
        print(card)
    
def show_all(player,dealer):
    # show all the dealer's cards
    # calculate and display value
    # show all the player's cards
    # calculate and display value 
    
    print("\n Player's hand:")
    for card in player.cards:
        print(card)
    print(f"Value of player's hand is: {player.value}")

# or you can use
    # * makes it apply to all items in list, sep is seperator
    # in this case seperator by new line 
    print("\n Dealer's hand: ",*dealer.cards,sep='\n')
    print(f"Value of dealer's hand is: {dealer.value}")
    
```
