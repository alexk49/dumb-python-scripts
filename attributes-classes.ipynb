```python
class Dog():
    
    def __init__(self,breed):
        
        self.breed = breed
```

```python
my_dog = Dog()
```

^ when you create a "Dog" object, python now expects a breed because of the assignment

```python
my_dog = Dog(breed = 'Lab')
```

```python
type(my_dog)
```

```python
my_dog.breed
```

Attributes
we take in the argument
assign it using self.attribute_name 


```python
class Dog():
    
    def __init__(self,mybreed):
        
        #Attributes
        # we take in the argument
        #assign it using self.attribute.name
        self.breed = mybreed
```

```python
my_dog = Dog(mybreed='Huskie')
```

```python
type(my_dog)
```

```python
Dog
```

```python
my_dog.breed
```

we only have the parameter "breed" as mybreed was just a parameter for the argument

by convention the parameter name will match the attribute and the parameter argument

we can have multiple arguement

```python
class Dog():
    def __init__(self,breed,name,spots):
        self.breed = breed
        self.name = name
        
        #expect boolean True/False
        self.spots = spots
```

```python
my_dog = Dog(breed='lab',name='Sammy',spots=False)
```

```python
my_dog.breed
```

```python
my_dog.name
```

```python
my_dog.spots
```
