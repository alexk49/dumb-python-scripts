Counter looks like a dictionary, but is a dictionary sub class. Still stored in dictionary keys, but the count in the object are stored as values

```python
test = "dhadha"
```

```python
Counter(test)
```

```python
a_test_string = 'the quick brown fox jumps over the lazy dog'
collection = Counter(a_test_string)
print(collection)
```

```python
sentence = "How many times does each word show up in this sentence?"
```

```python
Counter(sentence.split())
```

```python
Counter(sentence.lower().split())
```

```python
letters = 'aaaaaabbbbbbbbccccccccccddddddddddd'
```

```python
c = Counter(letters)
```

```python
c
```

```python
c.most_common()
```

```python
list(c)
```

Default dictionary will assign a value where a key error would have occured

```python
d = defaultdict(lambda: 0)
```

```python
d['correct'] = 100
```

```python
d['correct']
```

```python
d['Wrong key']
```
