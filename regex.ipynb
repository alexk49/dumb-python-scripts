Regular expression (regex) allow us to search for general patterns in text data 

for example, a simple email format can be user@email.com

we know in this case we're looking for a pattern "text" + @ + "text" + ".com"

the re library allows us to create specialised pattern strings and then search for matches within the text

the primary skill for regex is understanding the special syntax for these pattern strings

Phone number:
    (555)-555-5555

    r"(\d\d\d)-\d\d\d-\d\d\d\d"

the r, lets python know that this is an expresssion. 

\d stands for digit

includes formatting like () - 

general identifiers mixed with the exact pattern you want 

r"(\{3})-\d{3}-\d{4}"

Regular expressions part two

\d means a digit

for example:

file_\d\d 

would match:

file_25

\w alphanumeric 

for example: \w-\w\w\w would match A-b_1

\s means white space

for example a\sb\sc would match a b c 

\D = a non digit

for example \D\D\D would match ABC 

\W\W\W = non-alphanumeric,

for example \W\W\W\W\W would match *-+=)

\S = Non-whitespace,

for example \S\S\S\S would match Yoyo

```python
text = "My phone number is 408-555-1234"
```

```python
phone = re.search("408-555-1234",text)
```

```python
phone
```

```python
phone = re.search(r"\d\d\d-\d\d\d-\d\d\d\d",text)
```

```python
phone
```

remember r before the regex pattern, as the backslashes can be used for other things e.g \n for a new line 

```python
phone.group()
```

can use quantifiers as the number of repetitions of a particular character

" + " means occurs one of more times,

for example Version\w\w+ would match Version A_b1_1

{3} means exactly 3 times, \D{3} would match abc

{2,4} occurs 2 to 4 times, \d{2,4} would match 123 

{3,} occurs 3 or more, so \w{w,} anycharacters 

" * " occurs zero or more times,  A* B* C* would match AAACC

? once or none, plurals? would match plural

```python
phone = re.search(r'\d{3}-\d{3}-\d{4}',text)
```

```python
phone
```

extract just area codes. 

re.compile - indicates group of expressions, compiled into one. But it still understands they are seperate groupings that can be called 

the () indicate the groups

```python
phone_pattern = re.compile(r"(\d{3})-(\d{3})-(\d{4})")
```

```python
results = re.search(phone_pattern,text)
```

```python
results
```

```python
results.group()
```

```python
print(results[0])
print(results[1])
print(results[2])
print(results[3])
```

```python
print(results.group(0))
print(results.group(1))
print(results.group(2))
print(results.group(3))
```
